```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\Users\kharr\OneDrive\Desktop\vehicles original.csv')
df_sm = df[['year', 'make', 'model', 'co2TailpipeGpm', 'VClass', 'fuelType', 'eng_dscr']]
df_sm['co2taillbsmile'] = df_sm['co2TailpipeGpm'] *0.00220462
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicle 4WD')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicle 2WD')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicle')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicles')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicle/4wd')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicle/2wd')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicles/4wd')]
df_sm = df_sm[~(df_sm['VClass'] == 'Special Purpose Vehicles/2wd')]
df_sm = df_sm[~(df_sm['VClass'] == 'Vans Passenger')]
df_sm = df_sm[~(df_sm['VClass'] == 'Vans, Cargo Type')]
df_sm = df_sm[~(df_sm['VClass'] == 'Vans, Passenger Type')]
df_sm = df_sm[~(df_sm['VClass'] == 'Subcompact Cars')]
df_sm = df_sm[~(df_sm['VClass'] == 'Vans')]
df_sm['VClass'].isna().sum()
```

```python

HEV = df_sm[df_sm['eng_dscr'].str.match('HEV.*')== True]
PHEV = df_sm[df_sm['eng_dscr'].str.match('.*PHEV.*')== True]
EV = df_sm[df_sm['fuelType']=='Electricity']

df_sm['CO2HEV'] = np.where(df_sm['eng_dscr'].str.match('HEV.*')== True,4482.0,0)
df_sm['CO2PHEV'] = np.where(df_sm['eng_dscr'].str.match('.*PHEV.*')== True,5554.6,0)
df_sm['CO2EV'] = np.where(df_sm['fuelType']=='Electricity',7151.8,0)

total = df_sm['CO2HEV']+df_sm['CO2PHEV']+df_sm['CO2EV']
df_sm['CO2VM_COMBO'] = total

df_sm['CO2VM_COMBO']=df_sm['CO2VM_COMBO'].replace(0,4387.2)
df_sm.describe()
```

```python
#aggregate to remove duplicate make/models with different engine or drive properties. Average CO2lb/mile per make/model/year

aggr_all= df_sm.groupby(['year','make','model', 'VClass','fuelType', 'CO2VM_COMBO'])['co2taillbsmile'].sum().reset_index(name ='CO2lbmi')
aggr_all
```

```python
#Export

aggr_all.to_excel(r'C:\Users\kharr\OneDrive\Desktop\Vehicle Data Cleaned.xlsx', index = False)
```
